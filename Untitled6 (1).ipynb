```python
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.linear_model import LogisticRegression, LassoCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
import numpy as np

def hybrid_feature_selection(X, y, n_features_to_select=5):
    # Convert X to a DataFrame if it's not already one
    if not isinstance(X, pd.DataFrame):
        X = pd.DataFrame(X)

    # Step 1: Filter Methods
    filter_method = SelectKBest(score_func=chi2, k='all')
    X_filtered = filter_method.fit_transform(X, y)
    X_filtered = pd.DataFrame(X_filtered, columns=X.columns)

    # Step 2: Wrapper Methods
    wrapper_model = RandomForestClassifier()
    rfe = RFE(estimator=wrapper_model, n_features_to_select=10)
    X_rfe = rfe.fit_transform(X_filtered, y)
    selected_columns_rfe = X_filtered.columns[rfe.support_].tolist()

    # Step 3: Embedded Methods
    embedded_model = LassoCV(cv=5)
    embedded_model.fit(X_rfe, y)
    coef = embedded_model.coef_

    if coef.ndim == 1:
        mask = coef != 0
        selected_columns_embedded = [col for col, keep in zip(selected_columns_rfe, mask) if keep]
        X_embedded = X_rfe[:, mask]
    else:
        selected_columns_embedded = [col for col, keep in zip(selected_columns_rfe, coef != 0) if keep]
        X_embedded = X_rfe[:, coef != 0]

    # Check if the number of remaining features is greater than the desired number
    n_remaining_features = len(selected_columns_embedded)
    if n_remaining_features <= n_features_to_select:
        # If not, skip the hybrid method and use the features from the embedded method
        selected_features = selected_columns_embedded
        X_hybrid = X_embedded
    else:
        # Step 4: Hybrid Methods
        hybrid_model = LogisticRegression()
        rfe_hybrid = RFE(estimator=hybrid_model, n_features_to_select=n_features_to_select)
        X_hybrid = rfe_hybrid.fit_transform(X_embedded, y)

        # The final set of features
        selected_features = [selected_columns_embedded[i] for i, keep in enumerate(rfe_hybrid.support_) if keep]

    # Train the final model
    final_model = RandomForestClassifier()
    final_model.fit(X_hybrid, y)

    # Evaluate the model
    evaluation_results = evaluate_model(final_model, X_hybrid, y)

    return selected_features, evaluation_results, X.columns[X.columns.isin(selected_features)].tolist()

def evaluate_model(model, X_test, y_test):
    # Predict the responses for the test set
    y_pred = model.predict(X_test)

    # Calculate the accuracy score
    accuracy = accuracy_score(y_test, y_pred)

    # Calculate the confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)

    # Calculate the classification report
    class_report = classification_report(y_test, y_pred)

    return {'accuracy': accuracy, 'confusion_matrix': conf_matrix, 'classification_report': class_report}


selected_features, evaluation_results, selected_columns = hybrid_feature_selection(X, y, n_features_to_select=10)
print("Selected features:", selected_features)
print("Evaluation results:", evaluation_results)
print("Selected columns:", selected_columns)
```
