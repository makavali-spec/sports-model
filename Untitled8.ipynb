```python
#The feature selection using a Random Forest classifier has identified several important features. Here are the top 10 features based on their importance:

rf_columns = ['nrfi_yrfi','away_batter_6_runs_added_per_attempt_from_base_running',
              'home_pitcher_runs_allowed_below_avg_per_out',
              'away_pitcher_home_run_allowed_per_inning',
              'home_pitcher_wins_gained_above_avg_per_out',
              'home_batter_4_runs_added_per_attempt_from_avoiding_double_play',
              'away_pitcher_first_inning_on_base_percentage_allowed',
              'home_batter_1_runs_added_per_attempt',
              'home_batter_1_runs_added_per_attempt_from_batting',
              'home_batter_5_home_runs_per_attempt',
              'away_batter_5_runs_added_per_attempt_from_avoiding_double_play']

data_tr = df_tr_1[rf_columns]
data_tr
```

```python
data_tr.dropna(axis = 0, inplace=True)
```

```python
data_tr.isna().sum()
```

```python
data_tr
```

```python

from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

# Encode the target variable
label_encoder = LabelEncoder()
data_tr['nrfi_yrfi_encoded'] = label_encoder.fit_transform(data_tr['nrfi_yrfi'])

# Define features and target
X = data_tr.drop(['nrfi_yrfi', 'nrfi_yrfi_encoded'], axis=1)
y = data_tr['nrfi_yrfi_encoded']

# Feature selection using Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Get feature importances
feature_importances = pd.DataFrame(rf.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

# Select features based on importance
selector = SelectFromModel(rf, prefit=True)
selected_features = X.columns[selector.get_support()]

selected_features, feature_importances.head(10)
```

```python

from sklearn.model_selection import cross_val_score

# Select only the top features for model training
X_selected = X[selected_features]

# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Perform cross-validation and calculate accuracy
accuracy_scores = cross_val_score(rf_classifier, X_selected, y, cv=5, scoring='accuracy')
average_accuracy = accuracy_scores.mean()

average_accuracy
```

```python

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC

# Feature Engineering: Polynomial Features
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_poly = poly.fit_transform(X_selected)

# Define a pipeline for Random Forest with hyperparameter tuning
pipeline_rf = Pipeline([
    ('classifier', RandomForestClassifier(random_state=42))
])
parameters_rf = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}
grid_rf = GridSearchCV(pipeline_rf, parameters_rf, cv=5, scoring='accuracy')
grid_rf.fit(X_poly, y)
best_rf_accuracy = grid_rf.best_score_

# Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_accuracy = cross_val_score(gb_classifier, X_poly, y, cv=5, scoring='accuracy').mean()

# Support Vector Machine
pipeline_svm = Pipeline([
    ('classifier', SVC(random_state=42))
])
parameters_svm = {
    'classifier__C': [0.1, 1, 10],
    'classifier__kernel': ['linear', 'rbf']
}
grid_svm = GridSearchCV(pipeline_svm, parameters_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_poly, y)
best_svm_accuracy = grid_svm.best_score_

best_rf_accuracy, gb_accuracy, best_svm_accuracy
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Feature Engineering: Polynomial Features
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_poly = poly.fit_transform(X_selected)

# Data preprocessing: Scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_poly)

# Address class imbalance
oversampler = RandomOverSampler()
X_resampled, y_resampled = oversampler.fit_resample(X_scaled, y)

# Stratified k-fold cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Define pipelines for various models
pipeline_rf = Pipeline([
    ('feature_selection', SelectFromModel(RandomForestClassifier(random_state=42))),
    ('classifier', RandomForestClassifier(random_state=42))
])

pipeline_gb = Pipeline([
    ('classifier', GradientBoostingClassifier(random_state=42))
])

pipeline_svm = Pipeline([
    ('classifier', SVC(random_state=42))
])

# Define parameter grids for hyperparameter tuning
param_grid_rf = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}

param_grid_gb = {
    'classifier__n_estimators': [100, 200],
    'classifier__learning_rate': [0.01, 0.1, 0.5],
    'classifier__max_depth': [3, 5, 10]
}

param_grid_svm = {
    'classifier__C': [0.1, 1, 10],
    'classifier__kernel': ['linear', 'rbf']
}

# Perform GridSearchCV for each model
grid_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=skf, scoring='accuracy')
grid_rf.fit(X_resampled, y_resampled)
best_rf_accuracy = grid_rf.best_score_

grid_gb = GridSearchCV(pipeline_gb, param_grid_gb, cv=skf, scoring='accuracy')
grid_gb.fit(X_resampled, y_resampled)
best_gb_accuracy = grid_gb.best_score_

grid_svm = GridSearchCV(pipeline_svm, param_grid_svm, cv=skf, scoring='accuracy')
grid_svm.fit(X_resampled, y_resampled)
best_svm_accuracy = grid_svm.best_score_

# Print best accuracies
print("Best Random Forest Accuracy:", best_rf_accuracy)
print("Best Gradient Boosting Accuracy:", best_gb_accuracy)
print("Best SVM Accuracy:", best_svm_accuracy)

# Additional steps for model evaluation, feature importance, and interpretation can be added here
```
