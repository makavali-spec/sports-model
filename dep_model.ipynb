```python
#The feature selection using a Random Forest classifier has identified several important features. Here are the top 10 features based on their importance:

rf_columns = ['nrfi_yrfi','away_batter_6_runs_added_per_attempt_from_base_running',
              'home_pitcher_runs_allowed_below_avg_per_out',
              'away_pitcher_home_run_allowed_per_inning',
              'home_pitcher_wins_gained_above_avg_per_out',
              'home_batter_4_runs_added_per_attempt_from_avoiding_double_play',
              'away_pitcher_first_inning_on_base_percentage_allowed',
              'home_batter_1_runs_added_per_attempt',
              'home_batter_1_runs_added_per_attempt_from_batting',
              'home_batter_5_home_runs_per_attempt',
              'away_batter_5_runs_added_per_attempt_from_avoiding_double_play']

data_tr = df_merge[rf_columns]
data_tr
```

```python

# Define your mapping dictionary
mapping = {
    'nrfi': 0,
    'yrfi': 1
}

# Apply the mapping using .loc
data_tr.loc[:, 'nrfi_yrfi'] = data_tr['nrfi_yrfi'].map(mapping)

# Display the DataFrame after mapping
data_tr
```

```python
outliers = {}

for column in rf_columns:
    Q1 = data_tr[column].quantile(0.25)
    Q3 = data_tr[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define a mask for values outside the IQR
    mask = (data_tr[column] < (Q1 - 1.5 * IQR)) | (data_tr[column] > (Q3 + 1.5 * IQR))

    # Add the outliers to the dictionary
    outliers[column] = data_tr[mask]

# Print the outliers
for column, outlier_df in outliers.items():
    print(f"Outliers for {column}:")
    print(outlier_df)
```

```python

from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

# Encode the target variable
label_encoder = LabelEncoder()
data_tr['nrfi_yrfi_encoded'] = label_encoder.fit_transform(data_tr['nrfi_yrfi'])

# Define features and target
X = data_tr.drop(['nrfi_yrfi', 'nrfi_yrfi_encoded'], axis=1)
y = data_tr['nrfi_yrfi_encoded']

# Feature selection using Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Get feature importances
feature_importances = pd.DataFrame(rf.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

# Select features based on importance
selector = SelectFromModel(rf, prefit=True)
selected_features = X.columns[selector.get_support()]

selected_features, feature_importances.head(10)
```

```python

from sklearn.model_selection import cross_val_score

# Select only the top features for model training
X_selected = X[selected_features]

# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Perform cross-validation and calculate accuracy
accuracy_scores = cross_val_score(rf_classifier, X_selected, y, cv=5, scoring='accuracy')
average_accuracy = accuracy_scores.mean()

average_accuracy
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.ensemble import BalancedRandomForestClassifier
from imblearn.combine import SMOTEENN
from sklearn.feature_selection import SelectFromModel
from xgboost import XGBClassifier
import numpy as np

# Define your data preprocessing steps
preprocessing_steps = [
    ('polynomial_features', PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)),
    ('scaling', StandardScaler()),
    ('robust_scaling', RobustScaler())
]

# Define classifiers to try
classifiers = {
    'RandomForest': RandomForestClassifier(random_state=42),
    'GradientBoosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42)
}

# Define oversampling techniques to try
oversamplers = {
    'RandomOverSampler': RandomOverSampler(),
    'SMOTE': SMOTE(),
    'SMOTEENN': SMOTEENN()
}

# Define parameter grids for hyperparameter tuning
param_grids = {
    'RandomForest': {
        'classifier__n_estimators': [100, 200],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5]
    },
    'GradientBoosting': {
        'classifier__n_estimators': [100, 200],
        'classifier__learning_rate': [0.01, 0.1, 0.5],
        'classifier__max_depth': [3, 5, 10]
    },
    'XGBoost': {
        'classifier__n_estimators': [100, 200],
        'classifier__learning_rate': [0.01, 0.1, 0.5],
        'classifier__max_depth': [3, 5, 10]
    }
}

# Initialize variables to track best accuracy and parameters
best_accuracy = 0
best_params = {}

# Loop over preprocessing steps
for scaler_name, scaler in preprocessing_steps:
    # Apply preprocessing steps
    X_preprocessed = scaler.fit_transform(X_selected)

    # Loop over classifiers
    for classifier_name, classifier in classifiers.items():
        # Define pipeline with classifier
        pipeline = Pipeline([
            ('classifier', classifier)
        ])

        # Loop over oversamplers
        for oversampler_name, oversampler in oversamplers.items():
            # Address class imbalance
            X_resampled, y_resampled = oversampler.fit_resample(X_preprocessed, y)

            # Stratified k-fold cross-validation
            skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

            # Define parameter grid for hyperparameter tuning
            param_grid = param_grids[classifier_name]

            # Perform GridSearchCV
            grid = GridSearchCV(pipeline, param_grid, cv=skf, scoring='accuracy')
            grid.fit(X_resampled, y_resampled)
            best_accuracy = max(best_accuracy, grid.best_score_)
            best_params = grid.best_params_

            # Print best accuracy for each combination
            print(f"Preprocessing: {scaler_name}, Classifier: {classifier_name}, Oversampler: {oversampler_name}, Best Accuracy: {grid.best_score_}")

# Print overall best accuracy and parameters
print(f"Overall Best Accuracy: {best_accuracy}")
print(f"Overall Best Parameters: {best_params}")

# Additional steps for model evaluation, feature importance, and interpretation can be added here
```
